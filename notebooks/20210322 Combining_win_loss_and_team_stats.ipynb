```python
def clean_strings_of_special_characters(df, column):
    for index, row in df.iterrows():
        mask = (df['away_team'] == row['away_team']) & (df['Date '] == row['Date '])
        clean = unicodedata.normalize("NFKD", row['away_team'])
        df.loc[mask, 'away_team_clean'] = clean
    return df
```

```python
TEAM_COLUMN = 'Team'
KICK_STAT_COLUMN = 'KI�?'
ONE_PERCENTER_STAT_COLUMN = '1%s'

KICK_COLUMN_TO_BE_ADDED = 'kicks'
ONE_PERCENTER_COLUMN_TO_BE_ADDED = 'one-percenters'
def add_team_stats_to_afl_win_loss(df_team_stats, df_win_loss, column):
    for index, row in df_team_stats.iterrows():
        mask = df_win_loss[column] == row[TEAM_COLUMN]
        df_win_loss.loc[mask, KICK_COLUMN_TO_BE_ADDED + '_' + column] = row[KICK_STAT_COLUMN]
        df_win_loss.loc[mask, ONE_PERCENTER_COLUMN_TO_BE_ADDED + '_' + column] = row[ONE_PERCENTER_STAT_COLUMN]
    return df_win_loss
```

```python
df_new = add_team_stats_to_afl_win_loss(df, df_team, 'home_team')
```

```python
df_new = add_team_stats_to_afl_win_loss(df, df_team, 'away_team')
```
