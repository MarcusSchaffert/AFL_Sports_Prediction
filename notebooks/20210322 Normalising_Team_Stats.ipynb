```python
df = pd.read_csv('20210322 AFL_2020_team_stats_and_win_loss_trimmed.csv')
```

```python
df
```

```python
def normalise_column(df, column):
    df[column + '_normalise'] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())
    return df
```

```python
df.columns
```

```python
df = normalise_column(df, df.columns[0])
df = normalise_column(df, df.columns[1])
df = normalise_column(df, df.columns[2])
df = normalise_column(df, df.columns[3])
```

```python
mask = df[df.columns[0]].isna() | df[df.columns[1]].isna() | df[df.columns[2]].isna() | df[df.columns[3]].isna()
```
